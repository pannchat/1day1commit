```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        
```

```python
class NodeMgmt:
    def __init__(self, head):
        self.head = head
    
    def insert(self, value):
        self.current = self.head
        while True:
            if value < self.current.value:
                if self.current.left != None: # 왼쪽에 이미 브랜치가 있다면?
                    self.current = self.current.left # 브랜치 노드로 가서 비교를 시작
                else:
                    self.current.left =  Node(value) # 없으면 그냥 브랜치로 추가하기
                    break
            else:
                if self.current.right != None:
                    self.current = self.current.right
                else:
                    self.current.right = Node(value)
                    break
    def search(self,value):
        self.current = self.head
        while self.current:
            if self.current.value == value:
                return True
            elif value < self.current.value:
                self.current = self.current.left
            else:
                self.current = self.current.right
        return False
    
    def delete(self, value):
        searched = False
        self.current = self.head
        self.parent = self.head
        while self.current :
            if self.current.value == value:
                searched = True
                break
            elif value < self.current.value:
                self.parent = self.current
                self.current = self.current.left
            else:
                self.parent = self.current
                self.current = self.current.right
        if searched == False:
            return False
        
        # 현재 current 노드는 삭제할 노드를 가리키고 있다.
        # 그리고 삭제노드의 parent노드를 가지고 있다.
        #만약 Leaf 노드 이면?
        if self.current.left == None and self.current.right == None:
            if value < self.parent.value : # parent 의 왼쪽값을 삭제할 것인지 오른 쪽 값을 삭제할 것인지.
                self.parent.left = None
            else:
                self.parent.right = None
            del self.current
        elif self.current.left != None and self.current.right == None:
            if value < self.parent.value:
                self.parent.left = self.current.left
            else:
                self.parent.right= self.current.left
        elif self.current.left == None and self.current.right != None:
            if value < self.parent.value:
                self.parent.left = self.current.right
            else:
                self.parent.right = self.current.right
        elif self.current.left != None and self.current.right != None:
            if value < self.parent.value:
                self.change = self.current.right
                self.change_parent = self.current.right
                while self.change.left != None:
                    self.change_parent = self.change
                    self.change = self.change.left
                if self.change.right != None:
                    self.change_parent.left = self.change.right
                else:
                    self.change_parent.left = None
                self.parent.left = self.change
                self.change.left = self.change.left
                self.change.right = self.current.right
        else:
            self.change = self.current.right
            self.change_parent = self.current.right
            while self.change.left != None:
                self.change_parent = self.change
                self.change = self.change.left
            if self.change.right != None:
                self.change_parent.left = self.change.right
            else:
                self.change_parent.left = None
            self.parent.right = self.change
            self.change.left = self.current.left
            self.change.right = self.current.right
        
    
```
